# 5. What is the least number of items could you order from the menu to meet one day's nutritional requirements?

```python
def div_nzero(n, d):
    return n / d if d else 0
```

```python
div_nzero(100,0)
```

```python
item_list = []
item_amount = []

for i in df.index:
    f_list = []
    fat = int(div_nzero(100.0 , df.loc[i,'Total Fat (% Daily Value)']))
    satur = int(div_nzero(100.0 , df.loc[i,'Saturated Fat (% Daily Value)']))
    chole = int(div_nzero(100.0 , df.loc[i,'Cholesterol (% Daily Value)']))
    sod = int(div_nzero(100.0 , df.loc[i,'Sodium (% Daily Value)']))
    carbo = int(div_nzero(100.0 , df.loc[i,'Carbohydrates (% Daily Value)']))
    diet = int(div_nzero(100.0 , df.loc[i,'Dietary Fiber (% Daily Value)']))
    vita = int(div_nzero(100.0 , df.loc[i,'Vitamin A (% Daily Value)']))
    vitc = int(div_nzero(100.0 , df.loc[i,'Vitamin C (% Daily Value)']))
    calc = int(div_nzero(100.0 , df.loc[i,'Calcium (% Daily Value)']))
    iron = int(div_nzero(100.0 , df.loc[i,'Iron (% Daily Value)']))
    f_list.extend((fat,satur,chole,sod,carbo,diet,vita,vitc,calc,iron))
    
    print(i,".",df.loc[i,'Item'], "least amount", max(f_list) )
    item_list.append((df.loc[i,'Item']))
    item_amount.append((max(f_list)))
   
```

```python
item_amount
```

```python
least_df = pd.DataFrame(list(zip(item_list,item_amount)),columns=['Item','Least Amount'])
```
